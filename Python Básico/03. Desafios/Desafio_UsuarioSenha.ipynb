## DESAFIO - USUÁRIO E SENHA

Crie um programa com os elementos abaixo:

1. Solicita o nome de usuário e senha
2. Verifica se ambos estão corretos (Caso sim, exibe uma mensagem de sucesso)
3. Verifica se ambos estão corretos (Caso não, exibe uma mensagem de erro)
4. Mensagem de erro 1: Caso o usuário esteja errado
5. Mensagem de erro 2: Caso a senha esteja errado
6. Usuário e senha "corretos"podem ser definidos como variáveis dentro do próprio código

```python
usuarios = {
    "Pedro": "senha123",
    "Lucas": "senha456"
}

def menu_inicial():
    while True:
        try:
            escolha = int(input("Digite 1 para entrar na sua conta ou 2 para criar um novo usuário: "))
            if escolha == 1 or escolha == 2:
                return escolha
            else:
                print("Por favor, digite 1 ou 2.")
        except ValueError:
            print("Por favor, digite uma opção válida.")

def login():
    tentativas_usuario = 0
    while tentativas_usuario < 3:
        user_name = input("Digite o usuário: ")
        if user_name in usuarios:
            tentativas_senha = 0
            while tentativas_senha < 3:
                user_password = input("Digite a senha: ")
                if usuarios[user_name] == user_password:
                    print("Login realizado com sucesso!")
                    return True
                else:
                    print("Senha incorreta")
                    tentativas_senha += 1
            else:
                print("Número de tentativas da senha excedido.")
                return False
        else:
            print("Usuário não encontrado")
            tentativas_usuario += 1
    print("Número de tentativas de usuário excedido.")
    return False

def cadastro():
    tentativas_nome = 0
    while tentativas_nome < 3:
        user_name = input("Digite o nome do novo usuário: ")
        if user_name in usuarios:
            print("Usuário já existente!")
            tentativas_nome += 1
            continue
        
        user_name_validation = input("Digite novamente o nome do usuário para confirmar: ")
        if user_name != user_name_validation:
            print("Os nomes de usuário não são iguais.")
            tentativas_nome += 1
            continue
        
        tentativas_senha = 0
        while tentativas_senha < 3:
            user_password = input("Digite a senha: ")
            user_password_validation = input("Confirme a senha: ")
            if user_password == user_password_validation:
                usuarios[user_name] = user_password
                print("Usuário cadastrado com sucesso!")
                return True
            else:
                print("As senhas não são iguais.")
                tentativas_senha += 1
        
        # Se exceder tentativas de senha, volta para nome do usuário
        print("Número de tentativas de senha excedido. Voltando para o nome do usuário.")
        tentativas_nome += 1
    
    print("Número de tentativas de nome excedido. Tente novamente mais tarde.")
    return False

# Loop principal
while True:
    escolha = menu_inicial()
    if escolha == 1:
        login()
    elif escolha == 2:
        cadastro()

    continuar = input("Deseja fazer outra operação? (s/n): ").lower()
    if continuar != 's':
        print("Encerrando o programa.")
        break
```
