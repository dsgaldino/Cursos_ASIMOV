## DESAFIO DA ENTREVISTA

Crie um programa com os elementos abaixo:

1. Pede pela nome e idade
2. Dá um "oi" para a pessoa
3. Conta quantas letras possui o nome
4. Fala quantos anos a pessoa terá daqui a 5 anos.

```python
# Pergunta o nome
name = input("Qual o seu nome?")
```

```python
# Pergunta a idade
while True:
    try:
        age = int(input("Qual a sua idade? "))
        break
    except ValueError:
        print("Por favor, digite um número inteiro válido para a idade.")
```

```python
# Conversação

print("Olá " + name + ", seja ben-vindo!")
print("Seu nome possui " + str(len(name)) + " letras.")
print("Você terá " + str((age+5)) + " anos daqui a 5 anos")
```
