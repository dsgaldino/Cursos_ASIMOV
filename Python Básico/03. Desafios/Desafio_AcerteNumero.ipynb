## DESAFIO - ACERTE O NÚMERO

Crie um programa com os elementos abaixo:

1. Escolhe um número secreto.
2. Pede o usuário para adivinha o número
3. Indica se o usuário acertou o número
4. Se não acertar, fornece uma dica (número mais alto ou mais baixo)
5. Repete 3x.

```python
import random

def game(num_max):
    global tentativas, numero

    while tentativas < 3:
        user_guess = int(input(f"Digite um número entre 0 e {num_max}: "))

        if user_guess == numero:
            print("Parabéns, você acertou!")
            break
        else:
            tentativas += 1
            print("Que pena, você errou!\n")
            print("Essa foi sua tentativa nº: " + str(tentativas))

            if tentativas == 3:
                print("Número de tentativas excedido. O número secreto era", numero)
                break
            else:
                if user_guess > numero:
                    print("\nO número que deve acertar é menor do que o seu palpite!")
                else:
                    print("\nO número que deve acertar é maior do que o seu palpite!")

# Loop principal
while True:
    continuar = input("Deseja iniciar o jogo? (s/n): ").lower()
    if continuar == 's':
        try:
            nivel = int(input("Selecione o modo que quer jogar:\n1 - Fácil\n2 - Médio\n3 - Difícil\n"))
            if nivel == 1:
                num_max = 10
            elif nivel == 2:
                num_max = 50
            elif nivel == 3:
                num_max = 100
            else:
                print("Por favor, escolha uma opção válida.")
                continue

            numero = random.randint(0, num_max)
            tentativas = 0
            game(num_max)

        except ValueError:
            print("Por favor, digite uma opção válida.")
    elif continuar == 'n':
        print("Encerrando o programa.")
        break
    else:
        print("Por favor, digite 's' para sim ou 'n' para não.")
```
